```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats as st
from scipy.stats import skewnorm
from scipy.stats import norm, skew, kurtosis
import matplotlib.pyplot as plt
```

```python
#%matplotlib notebook
```

```python
# X-Axis ranges from -3 and 3 with 0.001 steps
gaussian_distribution = np.arange(-3, 3, 0.01)

left_skew_temp = np.array([-15.458, -15.235, -15.659, -15.1547, -15.239, -15.891, -15.2358, -15.594, -15.235, -15.2354, -14.265, -12.124, -15.456, -15.698, -14.256, -15.684, -15.865, -15.2384, -15.568, -14.2354, -15.0126])
left_skew = np.append(left_skew_temp, gaussian_distribution)

right_skew_temp = np.array([15.458, 15.235, 15.659, 15.1547, 15.239, 15.891, 15.2358, 15.594, 15.235, 15.2354, 14.265, 12.124, 15.456, 15.698, 14.256, 15.684, 15.865, 15.2384, 15.568, 14.2354, 15.0126])
right_skew = np.append(gaussian_distribution, right_skew_temp)
```

## PDF & CDF - https://www.youtube.com/watch?v=PYIjkw0HN1Q

```python
# Plot Normal Distribution with Mean=0 and Standard Deviation=1
print("Guassian OR Normal Distribution")
print("Mean   : ", gaussian_distribution.mean())
print("Median : ", np.median(gaussian_distribution))
print(st.mode(gaussian_distribution))
plt.plot(gaussian_distribution, norm.pdf(gaussian_distribution, 0, 1))
plt.axvline(gaussian_distribution.mean(), color='cyan')
plt.axvline(np.median(gaussian_distribution), color='green')
#plt.axvline(st.mode(gaussian_distribution[0]), color='red')
plt.legend(['Mean : {}'.format(gaussian_distribution.mean()),'Median : {}'.format(np.median(gaussian_distribution))], loc='lower right')
plt.show()
```

```python
# Left Skewed
print("\nLeft Skewed Distribution")
print("Mean   : ", left_skew.mean())
print("Median : ", np.median(left_skew))
print(st.mode(left_skew))
plt.plot(left_skew, norm.pdf(left_skew, 0, 1))
plt.axvline(left_skew.mean(), color='cyan', label='Mean')
plt.axvline(np.median(left_skew), color='red', label='Median')
plt.legend(['Mean : {}'.format(left_skew.mean()),'Median : {}'.format(np.median(left_skew))])
plt.show()
plt.savefig("Left_Skew.png", transparent=False)
```

```python
# Right Skewed
print("\nRight Skewed Distribution")
print("Mean   : ", right_skew.mean())
print("Median : ", np.median(right_skew))
print(st.mode(right_skew))
plt.plot(right_skew, norm.pdf(right_skew, 0, 1))
plt.axvline(right_skew.mean(), color='cyan')
plt.axvline(np.median(right_skew), color='red')
plt.legend(['Mean : {}'.format(right_skew.mean()),'Median : {}'.format(np.median(right_skew))])
plt.show()
```

### Cummulative Density Function : 

#### norm.cdf(x, 0, 1)
where <br>
x = continuous values <br>
0 = Mean <br>
1 = Standard Deviation <br>

<b> This Distribution here is the Standard Normal Distribution </b>

```python
plt.plot(x, norm.cdf(x, 0, 1))
```

## Probability Density Function : 

<b> norm.pdf(x, 0, 1) </b> <br>
where <br>
x = continuous values <br>
0 = Mean <br>
1 = Standard Deviation <br>

```python
plt.plot(x, norm.pdf(x, 0, 1))
```

### np.arange(start, stop, step) 
where <br>
start is the starting integer <br>
stop is the ending integer <br>
step is the interval between the integers <br>


So here we have given (5, 15, 0.01)

Means it will store all the numbers in the variable 'temp' <b> ranging from 5 to 15 with interval of 0.01 </b>

So it'll loop numbers from 5 and perform 5+0.01 till 15

```python
temp = np.arange(5, 15, 0.01)

print(temp)
```

### <b> plt.plot(temp, norm.pdf(temp, 0, 1)) </b>

<b> This line means that we are creating a Normal Distribution with Mean=0 and SD=1 </b>

Norm.pdf means Normal Distribution using Probability Density Function

<b> (temp, 0, 1) </b> means

temp is the variable containing all the values for distribution <br>
0 is the Value for Mean <br>
1 is the Value for SD <br>

```python
print("Mean    : ", temp.mean())
print("Meadian : " , np.median(temp))

plt.plot(temp, norm.pdf(temp, 0, 1))
```

```python
sns.displot(temp)
```

```python
mu, sigma = norm.fit(temp)

print("Mean : ", mu)
print("SD   : ", sigma)
```
